## (2) Create the image of a rotated sin using rotation of coordinates.

```python
x = np.linspace(0, 2 * np.pi, 200)
f = np.sin(x)

angle = np.radians(30)
rotationMatrix = np.transpose(np.array([[np.cos(angle), -np.sin(angle)],
                                         [np.sin(angle), np.cos(angle)]]))

dots = np.dot(rotationMatrix, np.array([x,f]))

fig, ax = plt.subplots()
ax.plot(x, f, 'b-', label='Seno normal')
ax.plot(dots[0], dots[1], 'r-', label=f'Seno rotacionado em 30 graus')

ax.legend()
plt.show()
```

## (3) Create the image of a gaussian.

```python
mean = 0
sigma = 1

x = np.linspace(mean - 3*sigma, mean + 3*sigma, 100)
y = np.exp(-(x - mean)**2 / (2*sigma**2)) / (sigma * np.sqrt(2*np.pi))

fig, ax = plt.subplots()
ax.plot(x, y, 'b-', label='Função Gaussiana')

ax.legend()
plt.show()
```
